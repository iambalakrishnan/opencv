```python
img = cv2.imread('images/shapes.png')
#cv2_imshow(img)
grey = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#Threshold function -> anything is lite make it while remaining black
ret, thresh = cv2.threshold(grey, 50,255,1)
#threshold minimum - 50
#threshold maximum - 255 - white
#cv2_imshow(thresh)

contours, h = cv2.findContours(thresh, 1,2)
#contours specify outer boundry of pixel values of shape
#cv2_imshow(thresh)
for contour in contours:
    approx = cv2.approxPolyDP(contour,0.01*cv2.arcLength(contour,True), True)
    #print(approx)
    #print(len(approx))
    n = len(approx)
    if n>9:
        #this is Circle
        cv2.drawContours(img,[contour],0,(0,255,0),3)
    elif n==3:
        cv2.drawContours(img,[contour],0,(0,255,0),3)
    elif n==4:
        #square
        cv2.drawContours(img,[contour],0,(0,0,255),3)
    elif n==6:
        #Hexogen
        cv2.drawContours(img,[contour],0,(0,0,255),3)

cv2_imshow(img)
```
